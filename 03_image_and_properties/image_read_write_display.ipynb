```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import scipy.misc as msc
from PIL import Image
# import dicom

%matplotlib inline
```

```python
# img = msc.imread('../images/dice.png') # Does the same thing as:
img = mpimg.imread('../images/dice.png')
```

```python
Image.open('../images/dice.png')
```

```python
fig_size = [16, 9]
plt.rcParams['figure.figsize'] = fig_size
plt.imshow(img);
```

```python
# Convert to greyscale
grey = Image.open('../images/dice.png').convert('L')
grey # NOTE that the alpha channel has been lost
```

```python
img_array = msc.fromimage(grey)
img_array
```

```python
grey_img = msc.toimage(img_array)
grey_img
```

```python
# NOTE: imshow 'introduces' color of its own
plt.imshow(grey_img);
```

```python
# To override this behavior
plt.imshow(grey_img, cmap = 'gray');
```

```python
# Save image
msc.imsave('../images/diceGrey.png', grey_img)
```
