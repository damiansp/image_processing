```python
from PIL import Image
from scipy.ndimage.filters import gaussian_laplace, laplace
from skimage.feature import canny
from skimage.filters.edges import prewitt, sobel
import numpy as np
import scipy.misc as msc
import scipy.ndimage
```

```python
img_path = '../../images/dice.png'
a = Image.open(img_path).convert('L')
a
```

```python
# Mean Filter

# Init 5x5 filter, with constant weigths
k = np.ones((5, 5))/ 25
b = scipy.ndimage.filters.convolve(a, k)
b = msc.toimage(b)
b
```

```python
k = np.ones((10, 10))/ 100
b = scipy.ndimage.filters.convolve(a, k)
b = msc.toimage(b)
b
```

```python
# Init 5x5 filter, with constant weigths
k = np.ones((10, 10)) / 10
b = scipy.ndimage.filters.convolve(a, k)
b = msc.toimage(b)
b
```

```python
# Median Filter

b = scipy.ndimage.filters.median_filter(a, 
                                        size = 5, 
                                        footprint = None,
                                        output = None, 
                                        mode = 'reflect', 
                                        cval = 0.0, 
                                        origin = 0)
b = msc.toimage(b)
b
```

```python
b = scipy.ndimage.filters.median_filter(a, 
                                        size = 20, 
                                        footprint = None,
                                        output = None, 
                                        mode = 'reflect', 
                                        cval = 0.0, 
                                        origin = 0)
b = msc.toimage(b)
b
```

```python
# Max Filter
b = scipy.ndimage.filters.maximum_filter(a, 
                                         size = 5, 
                                         footprint = None, 
                                         output = None,
                                         mode = 'reflect',
                                         cval = 0.0,
                                         origin = 0)
b = msc.toimage(b)
b
```

```python
# Min
b = scipy.ndimage.filters.minimum_filter(a, 
                                         size = 5, 
                                         footprint = None, 
                                         output = None,
                                         mode = 'reflect',
                                         cval = 0.0,
                                         origin = 0)
b = msc.toimage(b)
b
```

```python
# Sobel edge detector
b = sobel(a)
b = msc.toimage(b)
b
```

```python
# Prewitt edge detector
b = prewitt(a, mask = None)
b = msc.toimage(b)
b
```

```python
# Canny edge detector
a2 = msc.fromimage(a)
b = canny(a2, sigma = 2.0)
b = msc.toimage(b)
b
```

```python
# Laplace 2nd derivative filter
# Works best when edges are relatively sparse in the image
b = laplace(a, mode = 'reflect')
b = msc.toimage(b)
b
```

```python
# Gaussian-Laplace
b = gaussian_laplace(a, 1, mode = 'reflect')
b = msc.toimage(b)
b
```
