```python
from PIL import Image
import math
import numpy as np
import scipy.misc as msc
```

```python
balls = '../../images/balls.jpg'
```

```python
im = Image.open(balls).convert('L')
im
```

```python
im1 = msc.fromimage(im)
im1
```

```python
im2 = 255 - im1
im3 = msc.toimage(im2)
im3
```

```python
gamma = 1.5
b1 = im1.astype(float)
bmax = np.max(b1)
b2 = b1 / bmax
# compute gamma-correction exponent
b3 = np.log(b2) * gamma
# perform gamma correction
c = np.exp(b3) * 25
c = c.astype(int)
c = msc.toimage(c)
c
```
